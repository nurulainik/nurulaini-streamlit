#Visualization & Explanatory Analysis#

Question 1

1. Is there a trend in monthly sales growth over time, and are there certain months that show significant increases or decreases?

```python
order_dataset = pd.read_csv("orders_dataset.csv")
order_items_dataset = pd.read_csv("order_items_dataset.csv")

merged_data = pd.merge(order_dataset, order_items_dataset, on="order_id")
merged_data['order_purchase_timestamp'] = pd.to_datetime(merged_data['order_purchase_timestamp'])
merged_data['order_month'] = merged_data['order_purchase_timestamp'].dt.to_period('M')

monthly_sales = merged_data.groupby('order_month')['price'].sum()

plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o', color='green')
plt.title('Trend Penjualan Bulanan')
plt.xlabel('Bulan')
plt.ylabel('Total Penjualan')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Question 2
2. What is the distribution of payment methods in e-commerce transactions, and are there payment methods that are more dominant than others?

```python
order_payments = pd.read_csv("order_payments_dataset.csv")

payment_method_distribution = order_payments['payment_type'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(payment_method_distribution.index, payment_method_distribution.values, color='skyblue')
plt.title('Distribusi Metode Pembayaran')
plt.xlabel('Metode Pembayaran')
plt.ylabel('Jumlah Transaksi')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Question 3
3. What is the distribution of customer reviews based on product review scores in e-commerce platforms, and are there patterns or trends that can be identified from this distribution?

```python
order_reviews = pd.read_csv("order_reviews_dataset.csv")

customer_satisfaction = order_reviews['review_score'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
customer_satisfaction.plot(kind='bar', color='skyblue')
plt.title('Distribusi Ulasan Pelanggan Berdasarkan Skor Ulasan Produk')
plt.xlabel('Skor Ulasan')
plt.ylabel('Jumlah Ulasan')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Question 4
4. How do sellers perform on e-commerce platforms based on the number of products sold, and who are the top 10 sellers based on the number of products sold?

```python
order_items = pd.read_csv("order_items_dataset.csv")
sellers = pd.read_csv("sellers_dataset.csv")

merged_data = pd.merge(order_items, sellers, on="seller_id")

products_sold_per_seller = merged_data.groupby('seller_id')['product_id'].count().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
products_sold_per_seller.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Penjual Berdasarkan Jumlah Produk Terjual')
plt.xlabel('Penjual (seller_id)')
plt.ylabel('Jumlah Produk Terjual')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
